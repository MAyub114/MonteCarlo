# Monty Hall simulator

Use Monte Carlo simulation to check whether to stay with the initial door or switch.

```python
# Core libraries
import numpy as np
import random

# Data visualisation
import matplotlib.pyplot as plt

# Suppress warnings
import warnings
warnings.filterwarnings("ignore")
```

## Functions

```python
def run_MontyHall(number_of_trials):
    count_no_change = 0
    count_should_have_changed = 0
    for i in range(number_of_trials):
        door  = ['ferrari', 'goat', 'goat']
        first_choice = random.choice(door)   # first choice = ferrari or goat
        door.remove(random.choice(door))     # randomly remove one element from ['ferrari', 'goat', 'goat']
        second_choice = random.choice(door)  # second choice = random pick from the two elements remaining 
        # Picked the door with the Ferrari and did not switch
        if first_choice == 'ferrari' and first_choice == second_choice:
            count_no_change +=1
        # Picked the wrong door (goat) but switched and chose the Ferrari
        elif first_choice == 'goat' and second_choice == 'ferrari':
            count_should_have_changed +=1
    odds_ratio = (count_should_have_changed / number_of_trials) / (count_no_change / number_of_trials)
    return odds_ratio
```

## Run simulation

```python
num_of_trials = np.arange(1, 10000, 1)
results = []

for i in num_of_trials:
    result = run_MontyHall(i)
    results.append(result)
    
# Show results
plt.figure(figsize=(8, 8))
plt.plot(num_of_trials, results)
plt.xlabel('Number of Trials')
plt.ylabel('Odds ratio in favour of changing the first choice');
```
